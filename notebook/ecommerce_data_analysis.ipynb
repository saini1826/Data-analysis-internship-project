## 4. Exploratory Data Analysis (EDA)

```python
# Calculating total unit price country-wise and sorted to identify top contributors
country_unit_price = df.groupby("Country")["UnitPrice"].sum()
plt.figure(figsize=(12,4))
sns.barplot(x=country_unit_price.index, y=country_unit_price.values)
plt.xticks(rotation=95)
plt.title("Total Unit Price by Country",color="b")
plt.xlabel("Country",color="blue")
plt.ylabel("Total Unit Price",color="blue")
plt.show()
```

## This graph highlights that the United Kingdom is the dominant contributor to total unit price, marking it the company's key revenue market. Sales from other countries are comparatively lower, showing a high dependency on e few regions. This insight supports markit prioritization and data-driven business strategy decisions.

```python
# The bar chart helps compare country-wise total revenue and identify top and low-performing markets
country_sales_series = df.groupby("Country")["TotalSales"].sum()
plt.figure(figsize=(12,4))
sns.barplot(x=country_sales_series.index, y=country_sales_series.values, color="steelblue")
plt.xticks(rotation=95)
plt.title("Total Sales by Country",color="blue")
plt.xlabel("Country",color="blue")
plt.ylabel("Total Sales",color="blue")
plt.tight_layout()
plt.show()
```

## This graph shows that the United Kingdom generates the highest total sales, clearly dominating all other countries. Most remaining countires contribute relatively low sales, indicating a highly concentrated revenue distribution. This insight helps identify key markets and supports strategic focus on high-performing regions.

```python
# The bar chart helps compare Top 10 country-wise total revenue
top_countries = country_sales_series.sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x = top_countries.index, y = top_countries.values,color="teal")
plt.title("Top 10 countries by total sales")
plt.xticks(rotation=45)
plt.ylabel("Total Sales")
plt.xlabel("Country")
plt.show()
```

## This graph highlights that the United Kigndom overwhelmingly leads total sales among the top 10 conutries, contributing a significantly larger share than the rest. Other countries show comparatively moderates sales, indicating a steep drop after the top market. This insight reveals strong revenue conccentration and emphasizes the importance of maintaining dominance in key markets while exploring growth opportunities in secondary regions. 

```python
# The bar chart helps compare Top 10 most frequently purchased products
top_products = df["Description"].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_products.index, y=top_products.values, color="seagreen")
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Most Frequently Purchased Products", color="blue")
plt.xlabel("Product Description", color="blue")
plt.ylabel("Number of Purchases", color="blue")
plt.tight_layout()
plt.show()
```

## This graph identifies the top 10 most frequently purchased products, showing that a small number of items drive the majority of purchase frequency. This demand concentration highlights key products for inventory optimization, targeted promotions, and consistent revenue generation. 

```python
# Visualized the top 10 customers based on total sales to identify high-value customers
customer_sales = df.groupby("CustomerID")["TotalSales"].sum().head(10)
plt.figure(figsize=(10, 8))
sns.barplot(x=customer_sales.index.astype(str), y=customer_sales.values,color="steelblue")
plt.xticks(rotation=40)
plt.title("Top 10 Customers by Total Sales",color="red")
plt.xlabel("Customer ID",color="blue")
plt.ylabel("Total Sales",color="blue")
plt.tight_layout()
plt.show()
```

## This graph highlights that Customer 12347.0 is the hightest revenue-generating customer, followed by customer 12356.0 The clear gap between top customers and others indicates revenue concentration among a small customer segment. These insights support customer segmentation, retention strategies, and targeted high-value customer engagement.

```python
# Identified the top 10 stock codes based on total quantity sold to highlight best-selling products
top_stock_codes = df.groupby("StockCode")["Quantity"].sum().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_stock_codes.index, y=top_stock_codes.values,hue=top_stock_codes.index,palette="coolwarm",legend=False)
plt.xticks(rotation=45)
plt.title("Top 10 Stock Codes by Total Quantity",color="red")
plt.xlabel("Stock Code",color="blue")
plt.ylabel("Total Quantity Sold",color="blue")
plt.tight_layout()
plt.show()
```

## This graph highlights that a small number of stock codes account for the highest total quantity sold, indicating strong demand concentration in specific products. The leading stock codes significantly outperform others, suggesting fast-moving items that are critical for incentory planing. These insights support stock prioritization, demand forecasting, and efficient supply chain management.

```python
sales_over_time = df.groupby("InvoiceDate")["TotalSales"].sum()
plt.figure(figsize=(15, 7))
plt.plot(sales_over_time.index, sales_over_time.values,color="red")
plt.title("Total Sales Over Time",color="red")
plt.xlabel("Invoice Date",color="blue")
plt.ylabel("Total Sales",color="blue")
plt.show()
```

## This time-series graph shows significant fluctuations in total sales over time, with noticeable spikes and occasional negative values. The variability indicates irregular purchase patterns and the presence of returns or cancellations impacting net importance of monitoring sales trends over time for forecasting, anomaly detection, and business planing.

```python
quantity_over_time = df.groupby("InvoiceDate")["Quantity"].sum()
plt.figure(figsize=(15, 7))
plt.plot(quantity_over_time.index, quantity_over_time.values)
plt.title("Total Quantity Sold Over Time",color="blue")
plt.xlabel("Invoice Date",color="black")
plt.ylabel("Total Quantity")
plt.style.use("seaborn-v0_8")
plt.show()
```

## This time-series graph shows fluctuations in total quantity sold over time, with periodic spikes and occasional negative values. These variations suggest irregular ordering patterns and the impact of product returns or cancellations. This insight supports demand forecasting, inventory control, and operational planing.

```python
quantity_over_time = df.groupby("Quantity")["TotalSales"].sum()
plt.figure(figsize=(15, 7))
plt.plot(quantity_over_time.index, quantity_over_time.values)
plt.title("Total Quantity vs TotalSales")
plt.xlabel("Quantity")
plt.ylabel("TotalSales")
plt.grid(True)
plt.show()
```

## The graph shows a positive relationship between quantity sold and total sales, where bulk orders contribute significantly to revenue.
# . Zero and negative quantity values indicate the presence of returns or order cancellations affecting net sales.
# . Overall, the insight highlights the need to focus on high-volume customers and reducce returns to improve business performance.

```python
# This scatter plot analyzes the relationship between total quantity sold and average unit price across product using log scales to handle wide value range
product_summary = df.groupby("Description").agg(Total_Quantity=("Quantity", "sum"), Average_UnitPrice=("UnitPrice", "mean")).reset_index()
plt.figure(figsize=(12, 8))
sns.scatterplot(data=product_summary[product_summary["Total_Quantity"] > 0], x="Total_Quantity", y="Average_UnitPrice", alpha=0.6)
plt.title("Total Quantity Sold vs Average Unit Price per Product",color="green")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Average Unit Price")
plt.xscale("log") 
plt.yscale("log")
plt.show()
```

# . Higher quantities are mostly associated with lower to mid-range average unit prices.
# . As unit price increases, the quantity sold generally decreases, indicating price sensitivity.
# . This suggests that affordable pricing drives higher sales volume.

```python
# This pie chart show the proportion of total sales contributed by the top 10 customers compared to all remaining customers
customer_sales = df.groupby("CustomerID")["TotalSales"].sum().sort_values(ascending=False)
top_n = 10
customer_sales_top_n = customer_sales.head(top_n)
other_customers_sales_total = customer_sales.iloc[top_n:].sum()
pie_data_customer_sales = pd.concat([customer_sales_top_n, pd.Series({"Other Customers": other_customers_sales_total})])
plt.figure(figsize=(8, 8))
plt.pie(pie_data_customer_sales, labels=pie_data_customer_sales.index.astype(str), startangle=140)
plt.title("Total Sales by Top 10 Customers")
plt.axis("equal") 
plt.show()
```

## . A small group of top customers contributes the majority of total sales.
## . This indicates high revenue dependency on key customers.
